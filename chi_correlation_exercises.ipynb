```python
# Exercise 2.1: Does tenure correlate with monthly charges?

H^O: Tenure does not correlate with monthly charges
H^a: Tenure does correlate with monthly charges
```

```python
corr, p =stats.pearsonr(brighteyes['tenure'], brighteyes['MonthlyCharges'])
corr, p
```

```python
a = 0.05

if p < a:
    print('Reject the null hypothesis')
else:
    print('Fail to reject the null hypothesis')
    
p
```

```python
def find_r(df):
    corr, p = stats.pearsonr(df['tenure'], df['TotalCharges'])
    return corr, p
```

```python
find_r(brighteyes)
```

```python
def find_allr(df):
#     corr, p = stats.pearsonr(df['tenure'], df['TotalCharges'])
    return pd.Series(stats.pearsonr(df['tenure'], df['TotalCharges']), index=['r','p-value'])
            
```

```python
brighteyes.groupby(['InternetService', 'PhoneService']).apply(find_allr)

# Exercise 2.3 Solution: We can reject the null hypothesis that there is no linear relationship
# between Internet service and phone service types
```
